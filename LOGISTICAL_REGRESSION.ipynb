```python
#Data Cleaning
df1 = df.drop(['Cabin','PassengerId','Name','Ticket','Cabin'],axis = 1)
```

```python
df1.isnull().sum()
```

```python
df1['Age'] = df1['Age'].fillna(df1['Age'].median())
```

```python
char_cols = ['Embarked']
```

```python
df1[char_cols] = df1[char_cols].fillna(df1.mode().iloc[0])
```

**Label Encoding**

udes to convert categorical data variables to numerical form, it helps machine learning algorithm to undertand handling the data correctly

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df1['Survived'] = le.fit_transform(df1['Survived'])
df1['Pclass'] = le.fit_transform(df1['Pclass'])
df1['Sex'] = le.fit_transform(df1['Sex'])
df1['Age'] = le.fit_transform(df1['Age'])
df1['SibSp'] = le.fit_transform(df1['SibSp'])
df1['Parch'] = le.fit_transform(df1['Parch'])
df1['Fare'] = le.fit_transform(df1['Fare'])
df1['Embarked'] = le.fit_transform(df1['Embarked'])
```

```python
sns.heatmap(df1.corr())
```

```python
#x,y split
x = df1.drop(['Survived'],axis = 1)
y = df1['Survived']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit_transform(x)
```

```python
#x,y train,test,split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size = 0.3,random_state = 12)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
Log = LogisticRegression()
Log.fit(x_train,y_train)
```

```python
y_predict = Log.predict(x_test)
```

```python
#Evaluation metrics

from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,precision_score,f1_score
cm = confusion_matrix(y_test,y_predict)
print('Confusion',cm)
recall = recall_score(y_test,y_predict)
print('Recall',recall)
accuracy = accuracy_score(y_test,y_predict)
print('Accuracy',accuracy)
precision = precision_score(y_test,y_predict)
print('Precision',precision)
F1score = f1_score(y_test,y_predict)
print('F1 score',F1score)
```
