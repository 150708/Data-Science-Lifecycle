**DATA_CLEANING**
Helps in handling data,dealing with outliers,correcting errors,normalizing and feature engineering.

```python
df1 = df.drop(['Cabin','PassengerId','Name','Ticket','Cabin'],axis = 1)
```

Removes the columns 'Cabin', 'PassengerId', 'Name', and 'Ticket' from the DataFrame df.Creates a new DataFrame df1 with the remaining columns after the specified ones are dropped.Inorder to proceed with data cleaning we need to drop the columns which are irrelevant and not depent upon predicting variable

```python
df1['Age'] = df1['Age'].fillna(df1['Age'].median())
```

```python
char_cols = ['Embarked']
```

```python
df1[char_cols] = df1[char_cols].fillna(df1.mode().iloc[0])
```

char_cols contains the names of the columns which have NAN values within 30% data loss,hence we need to deal the data with fillna function.this used to fill the most frequent values in the missing place of each column.
