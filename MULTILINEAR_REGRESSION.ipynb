```python
#Data Cleaning
df1 = df.drop(['Cabin','PassengerId','Name','Ticket','Cabin'],axis = 1)
```

```python
df1.isnull().sum()
```

```python
df1['Age'] = df1['Age'].fillna(df1['Age'].median())
```

```python
char_cols = ['Embarked']
```

```python
df1[char_cols] = df1[char_cols].fillna(df1.mode().iloc[0])
```

**Label Encoding**

udes to convert categorical data variables to numerical form, it helps machine learning algorithm to undertand handling the data correctly

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df1['Survived'] = le.fit_transform(df1['Survived'])
df1['Pclass'] = le.fit_transform(df1['Pclass'])
df1['Sex'] = le.fit_transform(df1['Sex'])
df1['Age'] = le.fit_transform(df1['Age'])
df1['SibSp'] = le.fit_transform(df1['SibSp'])
df1['Parch'] = le.fit_transform(df1['Parch'])
df1['Fare'] = le.fit_transform(df1['Fare'])
df1['Embarked'] = le.fit_transform(df1['Embarked'])
```

```python
#x,y split
x = df1.drop(['Survived'],axis = 1)
y = df1['Survived']
```

```python
x
```

```python
y
```

```python
#x,y train,test,split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size = 0.3,random_state = 12)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
le = LinearRegression()
le.fit(x_train,y_train)
```

```python
y_predict = le.predict(x_test)
```

```python
#model evaluation
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mae = mean_absolute_error(y_test,y_predict)
print("MAE",mae)
```

```python
mse = mean_squared_error(y_test,y_predict)
print("MSE",mse)
```

```python
rmse = np.sqrt(mse)
print("RMSE",rmse)
```

```python
lrmse = np.log(rmse)
print("Log RMSE",lrmse)
```
